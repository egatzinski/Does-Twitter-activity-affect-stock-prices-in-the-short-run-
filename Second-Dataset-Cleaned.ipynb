```python
import pandas as pd
import numpy as np
import re
#returns a subset of the columns
data2 = pd.read_csv('SecondDatasetFull.csv', error_bad_lines = False)
```

```python
#Checking the companies with the most occurrences in the dataset
data2['company_names'].value_counts().head(6)
```

```python
#storing the top 6 firms in a variable
topfirms2 = ['Twenty-First Century Fox', 'Alphabet Inc.', 'Discovery', 'Netflix', 'M&T Bank Corporation', 'The Gap', 'Masco Corporation', 'Applied Materials']
```

```python
# making sure that the dataset only contains the top 6 companies
#resetting the index of the dataframe
data2 = data2[data2['company_names'].isin(topfirms2)].reset_index()
```

## Data pre-processing

```python
##making sure that there are no values which are null in the 'text' column
data2 = data2[data2['text'].notnull()]
```

```python
#data cleaning
data2 = data2.drop(columns = ['verified;;;;;;;;;;'])
```

```python
#creating two columns which check for the number of words in a tweet and number of letters in a tweet respectively
data2['WordsPerTweet'] = data2['text'].apply(lambda x : len(x.split(' ')))
data2['LettersPerTweet'] = data2['text'].apply(lambda x : len(x))
```

```python
#storing 'WordsPerTweet' and 'LettersPerTweet' in a list and adding them to the dataset
data2['tuple'] = list(zip(data2['company_names'], data2['timestamp']))
tweetsday = data2.groupby(['company_names', 'timestamp'])['text'].count().to_frame().reset_index()
tweetsday['tuple'] = list(zip(tweetsday['company_names'], tweetsday['timestamp']))
dictdays = dict(zip(tweetsday['tuple'], tweetsday['text']))
data2['Tweets in a day'] = data2['tuple'].apply(lambda x : dictdays[x])
data2 = data2.drop(columns = ['tuple'])
```
